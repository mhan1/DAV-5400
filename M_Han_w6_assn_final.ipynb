- Task 7 answer: Based on the appearance of the boxplots of the diamond price for each cut, I can draw that the price for each cut's interquartile ranges are overlapped. The 'Fair' cut has the highest median price, and the 'Ideal' cut has the lowest median price. The 'Premium' cut has the highest third quartile of the price. All of the price distribution for each cut is right-skewed, as the median is close to the first quartile.

The box extends from the Q1 to Q3 quartile values of the data,
with a line at the median (Q2). The whiskers extend from the edges
of box to show the range of the data. The position of the whiskers
is set by default to `1.5 * IQR (IQR = Q3 - Q1)` from the edges of the box.
Outlier points are those past the end of the whiskers.
