- Question 1) Which day of week has the most flight arrival delay?

```python
'''I am making "late_arrival" dataframe by extracting only the rows that has
ArrDealay is larger than 0, which is late arrival.'''
late_arrival = my_frame2.iloc[:][my_frame2.ArrDelay > 0]
```

```python
# I am displaying only the five rows of "late_arrival" dataframe.
late_arrival.head()
```

```python
'''To extract the counts of unique values in the "DayOfWeek" column,
I am making the column as one-dimensional Series first, and then check the
unique values in that Series, and then use "value_counts" function to find
the DayOfWeek that has the most counts.'''

obj1 = late_arrival['DayOfWeek']
series1 = pd.Series(obj1)
series1.head()
```

```python
# I am checking the unique values in the series1.
series1.unique()
```

```python
'''I am counting the frequencies of each unique values in series1, which is DayOfWeek
when Arrival Delay is larger than 0.'''
pd.value_counts(series1.values)
```

- Answer 1): Based on the output above, DayOfWeek 1 (Monday) has the most arrival delay.

- Question 4) Which airlines(Unique Carrier) is performing the best in terms of the flight delay?

```python
'''I am making "on_time" dataframe by extracting only the rows that has
ArrDealay is smaller than or equal to 0 and DepDelay is smaller than or equal to 0,
which represent "on time" flights'''
on_time = my_frame2.iloc[:][my_frame2.DepDelay <= 0][my_frame2.ArrDelay <=0]
```

```python
'''I am displaying only the first five rows in the on_time dataframe, which have
only the information that were "on time" flights'''
on_time.head()
```

```python
'''To extract the counts of unique values in the "UniqueCarrier" column,
I am making the column as one-dimensional Series first, and then check the
unique values in that Series, and then use "value_counts" function to find
the DayOfWeek that has the most counts.'''

obj3 = on_time['UniqueCarrier']
series3 = pd.Series(obj3)
series3.head()
```

```python
# I am checking the unique values in the series3.
series3.unique()
```

- Based on above output, there 14 unique airlines in the series3.

```python
'''I am counting the frequencies of each unique values in series3, which is UniqueCarrier
when the flights are "on time".'''
pd.value_counts(series3.values)
```

- Answer 4): Based on above output, "XE" airline(ExpressJet Airlines) has the most "on time" flights record.
